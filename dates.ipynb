```python
import pandas as pd
import plotly.express as px
from datetime import datetime
from pandas.tseries.offsets import MonthBegin
```

```python
crash = pd.read_csv('Washington_Crash.csv')
zippop = pd.read_csv('zippop.csv')
topratecounties = ['Stevens', 'Skamania', 'Adams', 'Lincoln', 'Ferry']
counties = ['King', 'Pierce', 'Spokane', 'Yakima', 'Snohomish']
```

```python
crash['date'] = pd.to_datetime(crash['crash_dt'])
#crash['month'] = pd.to_datetime(crash['accmon'])
#crash['year'] = pd.to_datetime(crash['year'])
crash['date_rounded'] = crash['date'].dt.normalize().map(MonthBegin().rollback)
march_2020 = pd.to_datetime('03/13/2020')
recent = crash[crash['date'] >= march_2020] #All dates after March 13th, 2020
recent_death = recent[recent['injury']==4]
recent_death.rename(columns={"co_char":"County Name"}, inplace =True)
recent_death['County Name'].value_counts()
# top five counties based on NUMBER are King, Pierce, Spokane, Yakima, Snohomish
#print(recent_death['date_rounded'])
```

```python
# # narrow down df to top five counties
specificcounties = recent_death[recent_death['County Name'].isin(counties)]
print(len(specificcounties))
#print(specificcounties)
```

```python
fig = px.histogram(recent_death, x = 'date_rounded')
fig.update_layout(bargap=0.1, xaxis_title='Month', yaxis_title='Number of Crashes', title='Fatal Crashes per Month in Washington State since COVID')
fig.show()
```

```python
# fix ratios
fig2= px.histogram(specificcounties, x = 'date_rounded', color = 'County Name')
fig2.update_layout(bargap=0.1, xaxis_title='Month', yaxis_title='Number of Crashes', title='Fatal Crashes per Month in Top Five Washington Counties since COVID')
fig2.show()
```
